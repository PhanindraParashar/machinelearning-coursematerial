```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('white')
sns.set_palette('pastel')
plt.rcParams['figure.figsize'] = (8,6)
```

```python
d = sns.load_dataset('tips')
```

```python
d.head()
```

### Visualisation of a variable distribution, by category

```python
sns.boxplot(x='day', y='total_bill', data=d);
```

#### Visualisation by two categories, one on the x-axis, the other as colour

```python
sns.boxplot(x='day', y='total_bill', hue='sex', data=d);
```

#### Further colouring of the x-axis category

```python
d['weekend'] = d.day.isin(['Sat', 'Sun'])
```

```python
sns.boxplot(x='day', y='total_bill', hue='weekend', data=d, dodge=False);
```

#### Visualisation by three categories

```python
sns.catplot(x='smoker', y='total_bill', hue='sex', col='time', kind='box', data=d);
```

### See the individual points

```python
sns.swarmplot(x='day', y='total_bill', data=d);
```

```python
sns.swarmplot(x='day', y='total_bill', hue='sex', data=d);
```

```python
sw_ax = sns.swarmplot(x='day', y='total_bill', hue='sex', data=d, dodge=True);
```

```python
sw_ax = sns.swarmplot(x='day', y='total_bill', hue='sex', dodge=True, palette=['.3','.3'], data=d)
sw_ax = sns.boxplot(x='day', y='total_bill', hue='sex', notch=False, fliersize=0, data=d)
handles, labels = sw_ax.get_legend_handles_labels()
sw_ax.legend(handles[:2], labels[:2]);
```

```python
sns.stripplot(x='day', y='total_bill', hue='sex', data=d);
```

```python
sns.stripplot(x='day', y='total_bill', hue='sex', dodge=True, data=d);
```

```python
sw_ax = sns.stripplot(x='day', y='total_bill', hue='sex', dodge=True, palette=['.3','.3'], data=d)
sw_ax = sns.boxplot(x='day', y='total_bill', hue='sex', notch=False, fliersize=0, data=d)
handles, labels = sw_ax.get_legend_handles_labels()
sw_ax.legend(handles[:2], labels[:2]);
```

### Show distributions

```python
sns.violinplot(x='day', y='total_bill', hue='sex', split=True, data=d);
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', split=True, inner='quartile', data=d);
```

```python
sw_ax = sns.swarmplot(x='day', y='total_bill', hue='sex', dodge=True, palette=['.3','.3'], data=d)
sw_ax = sns.violinplot(x='day', y='total_bill', hue='sex', inner=None, data=d)
handles, labels = sw_ax.get_legend_handles_labels()
sw_ax.legend(handles[:2], labels[:2]);
```

```python
sns.boxenplot(x='day', y='total_bill', data=d);
```

### Kernel density estimation

```python
dw1 = d[d.weekend]
dw2 = d[~d.weekend]
ax = sns.kdeplot(dw1.total_bill, dw1.tip, shade=True, shade_lowest=False, cbar=True, label='Weekend', color='b', alpha=.5)
ax = sns.kdeplot(dw2.total_bill, dw2.tip, shade=True, shade_lowest=False, cbar=True, label='Weekday', color='r', alpha=.5)
plt.legend();
```
